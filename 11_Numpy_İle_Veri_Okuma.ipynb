# Numpy ile veri okuma

NumPy modülü ile dosyalardan veri okumak mümkündür ve özellikle sayısal veriler üzerinde çalışıyorsanız çok etkilidir. NumPy, genellikle CSV (Comma Separated Values) gibi dosyalardan hızlı ve verimli bir şekilde veri okumak için kullanılır.

### 1. numpy.loadtxt(): 
Metin Dosyalarından Veri Okuma
Bu fonksiyon, genellikle düzenli olarak yapılandırılmış sayısal veriler içeren düz metin dosyalarını okumak için kullanılır.

```python
import numpy as np

# Veriyi 'dosya.txt' dosyasından yükle
dosya = 'C:/Users/ASUS/Desktop/Ders/Python_ile_veri_bilimi/iris.data.txt'
```

```python
veri=np.loadtxt(dosya,delimiter=',',usecols=(0, 1, 2, 3))
print(veri)
```

veri=np.loadtxt(dosya,delimiter=',',skiprows=4)

Eğer normal okuma yapmak istesek hata verecek çünkü  np.loadtxt() fonksiyonunun veriyi sayısal (float) olarak okumaya çalıştığı, ancak veri içinde bir dize (örneğin 'Iris-setosa' gibi) olduğu için bunu başaramadığı durumlarda ortaya çıkar. Yani, loadtxt() sadece sayısal verilerle çalışabilir ve veri setinizde sayısal olmayan değerler (dizeler/stringler) varsa bu hata meydana gelir.

# 2. numpy.genfromtxt():
Eksik Verilerle Başa Çıkma
Bu fonksiyon, loadtxt() fonksiyonuna benzer, ancak eksik verilerle başa çıkabilme yeteneğine sahiptir. Eğer dosyada bazı hücreler boşsa, bu hücreler için belirli bir değer atanabilir (örneğin, NaN).

```python
import numpy as np

# Eksik veriler içeren dosya
veri = np.genfromtxt('iris.data.txt', delimiter=',', dtype=float)

print(veri)
```

usecols, numpy.loadtxt() ve numpy.genfromtxt() gibi fonksiyonlarda kullanılan bir parametredir ve dosyadan hangi sütunların okunacağını belirtmek için kullanılır. Bu sayede, dosyadaki belirli sütunları seçip yalnızca bu sütunları yükleyerek bellekte yer ve işlem süresi tasarrufu sağlanabilir.

usecols Parametresinin Kullanımı:
Tip: usecols, bir tamsayı dizisi, liste veya tuple olabilir.
Amaç: Belirli sütunları seçerek sadece o sütunların yüklenmesini sağlar.

```python
import numpy as np

# İlk üç sütunu okur (0, 1, 2 numaralı sütunlar)
veri = np.loadtxt('iris.data.txt', delimiter=',', usecols=(0, 1, 2,3))

print(veri)
```

# Önemli Not:

Çoğu zaman okuyacağımız tabloda farklı türdeki veri sütunları yer alabilir. Bu durumda .loadtxt() yerine .genfromtxt() metodu kullanılır. Bu metodda dtype=None olarak belirtirsek program her sütundaki veri tipini otomatik olarak belirlemeye çalışarak ilerler. Numpy diziler aynı türde verilerden oluşmak durumundadır. Bu nedenle farklı türdeki verileri okuduğumuz zaman dizilerden oluşan bir dizi elde edilir. Ana diziyi oluşturan diziler verinin her satırını ifade eder. 

```python
import numpy as np
dosya='iris.data.txt'
veri = np.genfromtxt(dosya, delimiter='\t',dtype=None)
print(veri)
```

# CSV Dosyasını Okuma
CSV dosyasındaki verileri NumPy ile hızlı bir şekilde yükleyebilirsiniz. csv.reader() gibi yöntemler yerine NumPy'yi kullanarak verileri sayısal olarak işlemeniz daha kolay olabilir. recfromcsv gibi bir metod ilede okuma yapabilirsiniz

```python
veri = np.genfromtxt('iris.csv', delimiter=',', skip_header=1, dtype=float)
print(veri)
```

```python
veri=np.recfromcsv('iris.csv', delimiter='\t',  dtype=None)
print(veri)
```
