## Veri Nedir?

Veri, ham (işlenmemiş) gerçek bilgi parçacığına verilen addır.

Veriler ölçüm, sayım, anket, deney, gözlem, araştırma vb. gibi metodlarla toplanırlar.

Ölçüm veya sayım yolu ile ifade edilen verilere nicel veri, sayısal bir değer bildirmeyen verilere nitel veri denir.

## Veri Bilimi Nedir?

Veri bilimi, yapılandırılmış (veri tablolarına düzgünce sığan ve sayılar, kısa metin, tarihler) ve yapılandırılmamış verilerden ( boyutları veya yapıları nedeniyle veri tablosuna tam olarak sığmaz (örneğin ses ve video dosyaları ile büyük metin belgeleri)) bilgi ve öngörü elde etmek için bilimsel yöntemleri, süreçleri, algoritmaları ve sistemleri kullanan çok disiplinli bir alandır. Matematik, istatistik, bilgisayar bilimi ve bilgi bilimi bağlamından birçok teknik ve teori kullanır.

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
# Resmin URL'si
url = 'https://miro.medium.com/v2/resize:fit:720/format:webp/1*F6loO_Snt5mYbnbZDKuMdg.png'
```

```python
# Resmi indirme
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Resmi matplotlib ile gösterme
plt.imshow(img)
plt.axis('off')  # Eksenleri kapatma
plt.show()
```

# Numpy 

NumPy, Python programlama dili için bir bilimsel hesaplama kütüphanesidir. NumPy'nin en temel ve güçlü özelliği, çok boyutlu diziler ve matrisler ile verimli şekilde çalışmamızı sağlayan veri yapıları sunmasıdır. Bu kütüphane, büyük veri kümeleri üzerinde hızlı işlemler yapabilmek için optimize edilmiştir ve genellikle veri analizi, makine öğrenmesi ve bilimsel hesaplamalar için kullanılır.

Veri Bilimi ve Analizi: NumPy, pandas ve scikit-learn gibi veri analizi ve makine öğrenmesi kütüphanelerinin temelini oluşturur.
    
Görüntü İşleme: NumPy, görüntüleri matrisler olarak temsil edip işlemleri hızlı ve verimli bir şekilde yapmayı sağlar.
    
Bilimsel Hesaplamalar: NumPy, bilimsel ve mühendislik hesaplamalarında yaygın olarak kullanılır.

# Scipy

SciPy, Python dilinde bilimsel ve teknik hesaplamalar için kullanılan güçlü bir kütüphanedir. NumPy üzerine inşa edilmiştir ve onun veri yapılarıyla çalışır. SciPy, bilimsel hesaplama işlemleri için çok geniş bir fonksiyon yelpazesi sunar.

Lineer Cebir, İstatistiksel İşlemler, Optimize İşlemleri, Sayısal İntegrasyon, Sinyal İşleme, İnterpolasyon, Görüntü İşleme gibi birçok alanda işlem yapabilme yeteneğine sahiptir.

# Pandas

Pandas, Python programlama dili için geliştirilen, veri analizi ve veri işleme süreçlerini kolaylaştıran, güçlü ve esnek bir kütüphanedir. Pandas, özellikle veri manipülasyonu ve analizi için kullanılır ve büyük veri kümeleri üzerinde verimli çalışır. Excel benzeri tablolama işlemlerini Python ortamında yapmamıza olanak tanır.

Veri Yapıları, Veri Okuma ve Yazma, Veri Manipülasyonu, Veri Analizi, Zaman Serileri, Eksik Verilerle Çalışma gibi işlemleri gerçekleştirebilir.

# SciKit-Learn

Scikit-Learn, Python programlama dili için geliştirilmiş, makine öğrenmesi ve veri madenciliği uygulamaları için kullanılan popüler bir kütüphanedir. Kullanıcı dostu bir arayüze sahip olan Scikit-Learn, çeşitli makine öğrenmesi algoritmalarını, veri ön işleme araçlarını ve model değerlendirme tekniklerini içerir.

Denetimli Öğrenme, Denetimsiz Öğrenme, Veri Ön İşleme, Model Değerlendirme ve Seçimi, Hiperparametre Ayarlama gibi birçok işlemi gerçekleştirmek amacıyla kullanılır. 

# MatplotLib

Matplotlib, Python programlama dilinde veri görselleştirme için kullanılan en popüler kütüphanelerden biridir. Kullanıcıların çeşitli grafikler, diyagramlar ve görseller oluşturmasına olanak tanır. Matplotlib, bilimsel ve mühendislik hesaplamalarıyla ilgili verileri görselleştirmek için yaygın olarak kullanılır ve açık kaynaklıdır.

Çeşitli Grafik Türleri, Alt Grafikler, Etiketler ve Başlıklar, Etkileşimli Grafikleri Destekleme, Kümeleme ve Kategorik Verilerle Çalışma gibi yöntemlerde tercih edilirler.
